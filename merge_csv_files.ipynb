```python
# load in player fantasy stats data as player_df
player_df = pd.read_csv('data/league_data/player_data/Yahoo-406.l.12345-Players.csv')
player_df.head()
```

```python
player_df['Player ID'] = player_df['ID']
position_df = pd.DataFrame(player_df, columns = ['Player ID', 'Eligible Positions'])
position_df
```

```python
#Merge datasets
draft_df = pd.merge(draft_df, position_df, how = 'left', on='Player ID')
draft_df
```

```python
team_draft_df = draft_df.drop(columns = ['Team Name', 'Avg. Pick', 'Avg. Round', 'Avg. Cost', '% Drafted'], axis=1)
team_draft_df.info()
```

```python
four_rounds_df = team_draft_df[team_draft_df['Round'] <= 4]
four_rounds_df
```

```python
first_four_df = four_rounds_df.drop(columns = ['Pick', 'Round', 'Player ID', 'Last Name', 'First Name'], axis=1)

first_four_df.info()
```

```python
first_four_dict = pd.DataFrame(first_four_df.groupby(['Team ID'])['Eligible Positions'].apply(lambda grp: list(grp.values)).to_dict())
first_four_dict
```

```python
num_rb = []

for col in first_four_dict:
    num_rb_col = first_four_dict[col].value_counts()['RB']
    num_rb.append(num_rb_col)

first_four_dict.loc[len(first_four_dict)] = num_rb
first_four_dict = first_four_dict.T.reset_index()
first_four_dict.columns = ['ID', 'pick_1', 'pick_2', 'pick_3', 'pick_4', 'num_rb']
first_four_dict
```

```python
first_four_dict['strategy'] = first_four_dict['num_rb']
first_four_dict['strategy'] = first_four_dict['strategy'].replace(to_replace = (3, 4, 5), value = 2)

# Check value counts
first_four_dict['strategy'].value_counts()
```

```python
to_drop = ['pick_1', 'pick_2', 'pick_3', 'pick_4', 'num_rb']
strategy_df = first_four_dict.drop(to_drop, axis=1)
strategy_df
```

```python
#Merge datasets
team_strategy_df = pd.merge(team_df, strategy_df, how = 'left', on='ID')
team_strategy_df = team_strategy_df.drop(columns = ['Name', 'Commissioner', 'Manager', 'Image', 'Wins', 'Losses', 'Ties'], axis=1)
```
